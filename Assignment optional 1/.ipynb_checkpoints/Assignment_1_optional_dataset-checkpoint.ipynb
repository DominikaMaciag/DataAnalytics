```python
# Assignment 1 - optional dataset
# author: Dominika Maciąg 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Dataset that is being used comes from LendingClub - a financial services company

# Excercise 1
# reading csv file
# a) Load data from csv file (lets print first 5 rows)
df = pd.read_csv('lending-club-data.csv', low_memory=False)
df.head()
```

```python
# b) Print the variable list. 
# Select categorical varables:
categorical_df = list(df.columns.values)
categorical_df
```

```python
# Select numerical variables:
numerical_df = df.describe()
numerical_df
```

```python
# Summary: From computed basic statistics of numerical variables we can observe that not all of them are meaningful. 
# For example id, zip_code, member_id are not meaningful values - which means that computing statistics in those cases won't 
# give us useful result. For example we don't need median of numerical value of our users id's.
```

```python
# Excercise 2 
# Using scatter plot:

# a) Show the relation of 
x_axis = df.loc[:,"zip_code"]
y_axis = df.loc[:,"annual_inc"]
plt.scatter(x_axis,y_axis)
plt.ylabel("annual payment")
plt.xlabel("zipcode")
plt.show()
# Summary: 1) There is more wealthier people living in first half of given zip codes.
# 2) Those living in the second part of given zip codes are more likely to want to get the loan due to the lack of funds. 
# The second observation is showed on plot 2c) 
```

```python
# b) Check how total payment is influenced by loans grade. 
x_axis_2 = df.loc[:,"grade_num"];
y_axis_2 = df.loc[:,"total_pymnt"];
plt.scatter(x_axis_2,y_axis_2);
plt.xlabel("grade");
plt.ylabel("total payment");
plt.show();
# Summary: for loans the lower the grade, the higher is the total payment. 
# The more you paid off the loan the lower grade you will have. 
```

```python
# c) Check relationship between annual income and loan amount. 
x_axis = df.loc[:,"loan_amnt"];
y_axis = df.loc[:,"annual_inc"];
plt.scatter(x_axis,y_axis);
plt.ticklabel_format(style = 'plain')
plt.xlabel("loan amount");
plt.ylabel("annual income");
plt.show();
# Summary: most loans are taken by people that has annual income lower than 500 000 $.
# We can see that taking loans with annual income over 1 000 000 $ is very rare. 
```

```python
# Exercise 3
# a) Create boxplot of loans grade versus it's total payment. Compare with plot 2b).
df.boxplot(column="total_pymnt", by=['grade_num']);
plt.xlabel("grade");
plt.ylabel("total payment");
plt.show();
# Summary: we can see the same relationship as in plot 2b) with the addition of median and interquartile range.
```

```python
# b) Create mosaic plot of any two categorical variables.
# To create mosaic plot I used the following categorical variables: grade and sub grade.
# There is no relationship between those two variables.
from statsmodels.graphics.mosaicplot import mosaic
sub_grade_num = df["sub_grade_num"].to_numpy()
grade_num = df["grade_num"].to_numpy()
data = pd.DataFrame({'sub_grade_num': sub_grade_num,'grade_num':grade_num})
mosaic(data, ['sub_grade_num','grade_num'])
plt.show()
```

```python
# Lets filter the data: we want annual income lower than 1 000 000 $ and dti (debt to income ratio) < 15% (mean dti)
filtered_df = df.loc[(df["dti"]<15) & (df["annual_inc"]<1000000)];
filtered_df.describe()
# a) Compute the basic statistics fulfilling above constraints:
```

```python
# Lets try creating one of the previous plots but with filtered data. 
# With filtered data we should be able to see the relationship clearer.

# Check relationship between annual income and loan amount:
x_axis = filtered_df.loc[:,"loan_amnt"];
y_axis = filtered_df.loc[:,"annual_inc"];
plt.scatter(x_axis,y_axis);
plt.ticklabel_format(style = 'plain')
plt.xlabel("loan amount");
plt.ylabel("annual income");
plt.show();
# Now we can see more precisely that in this range of earnings plot is more dense in area that has annual income below 200 000$
# In this dataset the smaller is the data range the more precise outcome we will get.
```

```python
# Lets find customers with best loan installment to income ratio.
# creating new column with desired ratio:
pd.set_option('mode.chained_assignment', None)
filtered_df["ratio_installment_income"] = (filtered_df["installment"]/filtered_df["annual_inc"])
f_df = filtered_df.sort_values(by=["ratio_installment_income"])
f_df.head() #printing the first five rows of the filtered dataframe

# Summary: Loan installment to income ratio shows the loan eligibility of the customer.
# Now we get list of customers that we can trust and give loan to.
```
