```python
# a) Show the relation of square footage and price.
# We can see that the more sqft living the higher the price. 
# We can also see that the price range for the square footage of the apartments interior living space is quite huge.
data_frame = pd.read_csv('kc_house_data.csv');
x_axis = data_frame.loc[:,"sqft_living"];
y_axis = data_frame.loc[:,"price"];
plt.scatter(x_axis,y_axis);
plt.xlabel("sqft living");
plt.ylabel("price");
plt.show();
```

```python
# b) Check how price is influenced by apartment's grade. Find other variables correlated with price.
# We can observe that the higher the grade, the greater is the price range of the apartments in that grade.
x_axis_2 = data_frame.loc[:,"grade"];
y_axis_2 = data_frame.loc[:,"price"];
plt.scatter(x_axis_2,y_axis_2);
plt.xlabel("grade");
plt.ylabel("price");
plt.show();
```

```python
# c) Check how good apartments are distributed over the city. 
# The most expensive apartments are mostly located in following zipcode areas: 98000, 98025, 98100
# Apartments are not evenly ditributed, there are areas that have smaller density of houses.
# First, plot zipcode versus price
x_axis = data_frame.loc[:,"zipcode"];
y_axis = data_frame.loc[:,"price"];
plt.scatter(x_axis,y_axis);
plt.xlabel("zipcode");
plt.ylabel("price");
plt.show();
```

```python
# Use apartments coordinates (longitude and lattitude) 
# in order to show where apartments of good grade are located.
x_axis = data_frame.loc[:,"long"];
y_axis = data_frame.loc[:,"lat"];
a = [i for i in range(21613)]
colours = a;

plt.scatter(x_axis,y_axis, c=colours, cmap="viridis");
plt.xlabel("longitude");
plt.ylabel("lattitude");
plt.show();
```

```python
# Exercise 3
# a) Create boxplot of apartment's grade versus it's price. Compare with plot 2b).
# Compared to 2b) plot using boxplot enables easy observation of where is the median (green stripe) and interquartile range.
# It is also easier to see outliers on the boxplot.
data_frame.boxplot(column="price", by=['grade']);
plt.xlabel("grade");
plt.ylabel("price");
plt.show();
```

```python
# b) Create mosaic plot of any two categorical variables.
# To create mosaic plot I used the following categorical variables: bedrooms and bathrooms.
from statsmodels.graphics.mosaicplot import mosaic
bedrooms = data_frame["bedrooms"].to_numpy()
bathrooms = data_frame["bathrooms"].to_numpy()
data = pd.DataFrame({'bedrooms':bedrooms,'bathrooms':bathrooms})
mosaic(data, ['bedrooms','bathrooms'])
plt.show()
```

```python
# Exercise 4
# "I am interested in the house with 3 bedrooms, 2 bathrooms and 2 floors"
# We need to filter data frame:
filtered_df = data_frame.loc[(data_frame["bedrooms"]==3.0) & (data_frame["bathrooms"]==2.0) & (data_frame["floors"]==2.0)];

# a) Compute the basic statistics of apartments fulfilling above constraints.
fd = filtered_df.describe();
print(fd)
```
